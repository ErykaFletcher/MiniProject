## custom function

```python
def simpleAdd(x,y):
    """
    a function to add 2 numbers, x and y
    """
    return (x+y)
```

```python
simpleAdd(8, 10)
```

```python
def compare(x,y):
    """
    This is a function to compare 2 number, x and y
    """
    #if statement
    if x>y:
        print(f"{x} is greater than {y}")
    elif x<y:
        print(f"{x} is smaller than {y}")
    else:
        print(f"{x} is the same as {y}")
```

```python
compare(4,5)
```

```python
import matplotlib.pyplot as plt
```

```python
#create a ablack 20 image (grayscale image)
black_img=np.zeros((200,200), dtype=np.uint8)

plt.imshow(black_img, cmap=plt.cm.gray)
plt.title("black grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#create a white image
white_img=255*np.ones((200,200))
white_img=np.uint8(white_img)

plt.imshow(white_img, cmap=plt.cm.gray, vmin=0, vmax=256)
plt.title("white grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a random color image
random_color_img = np.random.randint(0, high=256, size=(200,200,3), dtype="uint8")

plt.imshow(random_color_img)
plt.title("random color")
plt.xticks([]), plt.yticks([])
plt.show()
```
